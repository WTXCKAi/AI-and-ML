```python
def yield_accu_label(path):
    with open(path, "r", encoding="utf-8") as fp:
        lines = fp.readlines()
    for line in lines:
        dat = json.loads(line)
        yield dat['meta']['accusation'][0]


def init_data(path):
    accu, accuname = {}, {}
    accu_list = set(yield_accu_label(path))
    for index, accu_name in enumerate(accu_list):
        accuname[index] = accu_name.strip()
        accu[accu_name.strip()] = index
    return accu, accuname


def get_accu_class_num(accu):
    return len(accu)


def get_accu_name(accuname, index):
    return accuname[index]


def get_class_label(line, accu):
    return accu[line['meta']['accusation'][0]]


def read_train_data(path, accu):
    all_text, accu_label = [], []
    with open(path, "r", encoding="utf-8") as fp:
        lines = fp.readlines()
    for line in lines:
        dat = json.loads(line)
        all_text.append(dat["fact"])
        accu_label.append(get_class_label(dat, accu))
    return all_text, accu_label


def cut_text(all_text):
    cut = thulac.thulac(seg_only=True)
    train_text = []
    for text in all_text:
        train_text.append(cut.cut(text, text=True))
    return train_text

# 结巴分词，添加停用词
STOPWORDS_LOC = "./stopwords.txt"


def load_stopwords(stopwords_path):
    """ load stopwords into set from local file """
    stopwords = set()
    with open(stopwords_path, "r", encoding="utf-8") as fp:
        for line in fp.readlines():
            stopwords.add(line.strip())
    return stopwords


def cut_line(line, stopwords=None):
    # remove the date and time
    line = re.sub(r"\d*年\d*月\d*日", "", line)
    line = re.sub(r"\d*[时|时许]", "", line)
    line = re.sub(r"\d*分", "", line)

    word_list = jieba.cut(line)

    stopwords = load_stopwords(STOPWORDS_LOC)

    # remove the stopwords
    words = []
    for word in word_list:
        if word not in stopwords:
            words.append(word)

    text = " ".join(words)
    # correct some results
    # merge「王」and「某某」into「王某某」
    text = re.sub(" 某某", "某某", text)

    # merge「2000」and「元」into「2000元」
    text = re.sub(" 元", "元", text)
    text = re.sub(" 余元", "元", text)

    text = re.sub("价 格", "价格", text)
    return text


def jieba_cut_text(all_text):
    train_text = []
    for text in all_text:
        train_text.append(cut_line(text))
    return train_text
```

### 训练SVM模型

```python
def train_tfidf(train_data, dim=5000):
    tfidf = TFIDF(
        min_df=5,
        max_features=dim,
        ngram_range=(1, 3),
        use_idf=1,
        smooth_idf=1
    )
    tfidf.fit(train_data)
    return tfidf


def train_SVC(vec, label):
    SVC = LinearSVC()
    SVC.fit(vec, label)
    return SVC
```

```python
accu, accuname = init_data("./data_train.json")
all_text, accu_label = read_train_data("./data_train.json", accu)
```

```python
label = set(accu_label)
d = {}
for i in label:
    d[i] = accu_label.count(i)
```

```python
data = pd.DataFrame()
data["label"] = d.keys()
data["value"] = d.values()
```

```python
data.head()
```

```python
# 可视化训练集相关label的，fact数量
sns.barplot(x="label", y="value", data=data[:25])
```

```python
%%time
train_data = jieba_cut_text(all_text)
tfidf = train_tfidf(train_data)
vec = tfidf.transform(train_data)
svm_model = train_SVC(vec, accu_label)
```

### 保存训练好的模型

```python
def save_model():
    joblib.dump(tfidf, "tfidf.model")
    joblib.dump(svm_model, "svm_model.model")
```

```python
save_model()
```

### 测试模型

```python
class Predictor:
    def __init__(self):
        self.tfidf = joblib.load("tfidf.model")
        self.accu = joblib.load("eclf.model")
        self.cut = cut_line
    def predict_accu(self, vec):
        y = self.accu.predict(vec)
        return y[0]

    def predict(self, content):
        fact = self.cut(content)
        vec = self.tfidf.transform([fact])   
        return self.predict_accu(vec)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
```

```python
train_data = jieba_cut_text(all_text)
tfidf = joblib.load("tfidf.model")
vec = tfidf.transform(train_data)
```

```python
%%time
clf1 = LinearSVC()
clf2 = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial')
clf3 = RandomForestClassifier(n_estimators=100)
eclf = VotingClassifier(estimators=[
                    ('svm', clf1), ('lr', clf2), ('rf', clf3)], voting='hard'
                        )
eclf = eclf.fit(vec, accu_label)
```
